```python
import SimpleITK as sitk
import numpy as np
import matplotlib.pyplot as plt

def dice_coefficient(mask1_path, mask2_path):
    # Carga las imágenes usando SimpleITK
    mask1 = sitk.ReadImage(mask1_path)
    mask2 = sitk.ReadImage(mask2_path)
    
    # Convierte las imágenes a matrices numpy
    mask1_np = sitk.GetArrayFromImage(mask1)
    mask2_np = sitk.GetArrayFromImage(mask2)
    
    # Calcula el coeficiente de Dice
    intersection = np.sum(mask1_np * mask2_np)
    total_voxels = np.sum(mask1_np) + np.sum(mask2_np)
    
    dice = 2.0 * intersection / total_voxels if total_voxels > 0 else 0.0
    return dice

# Lista de rutas a los bundles
bundle_paths_1 = ['BPS_right_normal.nii.gz',
                  'BPS_left_normal.nii.gz',
                  'Fornix_normal.nii.gz',
                 'SLF_left_normal.nii.gz',
                 'SLF_right_normal.nii.gz',
                 'CP_normal.nii.gz',
                 'Cingulum_left_normal.nii.gz',
                 'Cingulum_right_normal.nii.gz']  # Primer conjunto de rutas

bundle_paths_2 = ['BPS_right_predicted.nii.gz',
                  'BPS_left_predicted2.nii.gz',
                  'Fornix_predicted.nii.gz',
                 'SLF_left_predicted2.nii.gz',
                 'SLF_right_predicted.nii.gz',
                 'CP_predicted.nii.gz',
                 'Cingulum_left_predicted.nii.gz',
                 'Cingulum_right_predicted.nii.gz']  # Segundo conjunto de rutas

dice_scores = []  # Almacenar los coeficientes de Dice
bundle_names = []  # Almacenar los nombres de los bundles

# Calcular los coeficientes de Dice y almacenar los nombres de los bundles
for i in range(min(len(bundle_paths_1), len(bundle_paths_2))):
    dice_score = dice_coefficient(bundle_paths_1[i], bundle_paths_2[i])
    dice_scores.append(dice_score)
    bundle_names.append(f'{bundle_paths_1[i]} - {bundle_paths_2[i]}')

# Graficar los coeficientes de Dice
plt.figure(figsize=(8, 6))
plt.barh(bundle_names, dice_scores, color='skyblue')
plt.xlabel('Dice Score')
plt.ylabel('Bundle')
plt.title('Coeficientes de Dice entre Bundles')
plt.grid(axis='x')
plt.show()
```
